```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create sample data instead of loading from a non-existent file
# This creates a DataFrame with sample sales data
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03'],
    'Product': ['Product A', 'Product B', 'Product A', 'Product C', 'Product B'],
    'Quantity Sold': [10, 5, 8, 12, 7],
    'Revenue ($)': [100, 75, 80, 180, 105]
}
df = pd.DataFrame(data)

# 2. Calculate the total revenue
total_revenue = df['Revenue ($)'].sum()

# 3. Find the best-selling product
best_selling = df.groupby('Product')['Quantity Sold'].sum().idxmax()
best_selling_qty = df.groupby('Product')['Quantity Sold'].sum().max()

# 4. Identify the day with the highest sales (by revenue)
highest_sales_day = df.groupby('Date')['Revenue ($)'].sum().idxmax()

# 5. Save the results to sales_summary.txt
with open('sales_summary.txt', 'w') as file:
    file.write(f"Total Revenue: ${total_revenue:,.2f}\n")
    file.write(f"Best-Selling Product: {best_selling} ({best_selling_qty} units sold)\n")
    file.write(f"Highest Sales Day: {highest_sales_day}\n")

# 6. Print the results in a user-friendly format
print("===== SALES SUMMARY =====")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Best-Selling Product: {best_selling} ({best_selling_qty} units sold)")
print(f"Highest Sales Day: {highest_sales_day}")

# 7. Bonus: Visualize sales trends
plt.figure(figsize=(8, 5))
df.groupby('Date')['Revenue ($)'].sum().plot(marker='o', linestyle='-')
plt.title('Daily Sales Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
